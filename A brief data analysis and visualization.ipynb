# 3. Understanding the Data

Reach out the data for some basic answers from the data base. Here, Country and Region with highest terrorist attacks and Country with highest Casualities are calculated.

```python
Terrorist_df[['country_txt','region_txt','provstate','nkill']].head(20)
```

```python
Total_Kills=len(Terrorist_df.groupby('country_txt').median().nkill)
Total_Kills
```

```python
Max_kill_in_top_10_countries=pd.DataFrame(Terrorist_df.groupby('country_txt').median().nkill)
Max_kill_in_top_10_countries.sort_values('nkill',ascending=False,inplace=True)
Max_kill_in_top_10_countries['Kill Percentage']=(Max_kill_in_top_10_countries.nkill/len(Terrorist_df.groupby('country_txt').median().nkill))*100
Max_kill_in_top_10_countries.rename(columns={'nkill':'No of Kills'},inplace=True)
Max_kill_in_top_10_countries.index.name='Countries'
Max_kill_in_top_10_countries.head(10)
```

```python
Max_kill_in_top_10_regions=pd.DataFrame(Terrorist_df.groupby('region_txt').median().nkill)
Max_kill_in_top_10_regions.sort_values('nkill',ascending=False,inplace=True)
Max_kill_in_top_10_regions['Kill Percentage']=(Max_kill_in_top_10_regions.nkill/len(Terrorist_df.groupby('region_txt').median().nkill))*100
Max_kill_in_top_10_regions.rename(columns={'nkill':'No of Kills'},inplace=True)
Max_kill_in_top_10_regions.index.name='Regions'
Max_kill_in_top_10_regions.head(10)
```

```python
Max_kill_in_top_10_states=pd.DataFrame(Terrorist_df.groupby('provstate').median().nkill)
Max_kill_in_top_10_states.sort_values('nkill',ascending=False,inplace=True)
Max_kill_in_top_10_states['Kill Percentage']=(Max_kill_in_top_10_states.nkill/len(Terrorist_df.groupby('provstate').median().nkill))*100
Max_kill_in_top_10_states.rename(columns={'nkill':'No of Kills'},inplace=True)
Max_kill_in_top_10_states.index.name='Provision/States'
Max_kill_in_top_10_states.head(10)
```

**4. Data Visualization - Terrorism around globe**

The below plot visualizes the no. of terrorist attacks in every year from 1970 - 2016.
It is clear that 2011-2016 are the worst affected years. Let's move on.

```python
plt.subplots(figsize=(15,6))
sns.countplot('iyear',data=Terrorist_df,palette='RdYlGn_r',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=90)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

1.The below plot visualizes the no. of casualities in every year from 1970-2016.

2.In the years 2012-2016, on an average 50000 people are victimised in terroristic activities.

3.In 2015, the toll of casualties reached above 70000 making the most deadlist year in the history.

```python
Year_Casualities_df = (Terrorist_df.groupby(['iyear']).sum())
print ("Year with Highest Number of Casualities : ",Year_Casualities_df.sort_values(by=['casualities'],ascending=[False])
       .index[0])

plt.subplots(figsize=(15,6))
sns.barplot(x = Terrorist_df['iyear'].unique(), y = Year_Casualities_df['casualities'], data= Year_Casualities_df, palette='rainbow')
plt.xticks(rotation=90)
plt.title('Terror Casualities Year Wise')
plt.show()
```

Year with Highest Number of Casualities :  2015

Terror Groups with more activity from 1970-2016, Talibans were the most active followed by ISIL.
